```python
# Week 1: Data Simulation for AI-Powered Green Energy Grid
This notebook generates a simulated dataset for:
- Waste (tons/day)
- COâ‚‚ levels (ppm)
- Energy demand (kWh)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate 30 days of simulated data
dates = pd.date_range(start="2025-01-01", periods=30, freq="D")

data = pd.DataFrame({
    "date": dates,
    "waste": np.random.randint(50, 150, size=30),   # tons
    "co2": np.random.randint(300, 500, size=30),    # ppm
    "energy_demand": np.random.randint(100, 300, size=30)  # kWh
})

# Save dataset to folder
import os
os.makedirs("data", exist_ok=True)
data.to_csv("data/simulated_data.csv", index=False)

data.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data["date"], data["waste"], label="Waste (tons)", marker='o')
plt.plot(data["date"], data["energy_demand"], label="Energy Demand (kWh)", marker='s')
plt.xlabel("Date")
plt.ylabel("Values")
plt.title("Simulated Waste & Energy Demand over Time")
plt.legend()
plt.grid(True)
plt.show()
```
